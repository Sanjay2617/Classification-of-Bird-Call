### Multi Class Classification 

> Class imbalance was handled by downloading additional data using a web scraping script from the Xeno-Canto website.

```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.utils import shuffle
import tensorflow as tf
from tensorflow.keras import layers, models, callbacks
import pandas as pd


def train_bird_cnn(epochs=20, batch_size=32, early_stopping_patience=3):
    with h5py.File('bird_spectrograms_3sec_native_fixed.h5', 'r') as f:
        X1 = np.transpose(f['rewbla'][:], (2, 0, 1)).astype('float32') 
        X2 = np.transpose(f['daejun'][:], (2, 0, 1)).astype('float32') 

    X = np.concatenate([X1, X2])
    y = np.array([0]*len(X1) + [1]*len(X2))
    X = X[..., np.newaxis]
    X, y = shuffle(X, y, random_state=42)

    X_temp, X_test, y_temp, y_test = train_test_split(X, y, test_size=0.2, stratify=y)
    X_train, X_val, y_train, y_val = train_test_split(X_temp, y_temp, test_size=0.2, stratify=y_temp)

    model = models.Sequential([
        layers.Input(shape=X_train.shape[1:]),
        layers.Conv2D(32, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(64, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.Conv2D(128, (3, 3), activation='relu'),
        layers.MaxPooling2D((2, 2)),
        layers.GlobalAveragePooling2D(),
        layers.Dense(128, activation='relu'),
        layers.Dropout(0.4),
        layers.Dense(1, activation='sigmoid')
    ])
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

    early_stop = callbacks.EarlyStopping(patience=early_stopping_patience, restore_best_weights=True)

    history = model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=epochs,
        batch_size=batch_size,
        callbacks=[early_stop],
        verbose=1
    )

    loss, acc = model.evaluate(X_test, y_test, verbose=0)
    y_pred_probs = model.predict(X_test, verbose=0)
    y_pred = (y_pred_probs > 0.5).astype(int)

    cm = confusion_matrix(y_test, y_pred)
    tn, fp, fn, tp = cm.ravel()

    report = classification_report(y_test, y_pred, target_names=["Red-winged Blackbird", "Dark-eyed Junco"])

    print(f"\n Model (epochs={epochs}, batch_size={batch_size}, patience={early_stopping_patience})")
    print(f"Test Accuracy: {acc:.4f}")
    print("Confusion Matrix:")
    print(cm)
    print(f"True Positives: {tp}, True Negatives: {tn}")
    print(f"False Positives: {fp}, False Negatives: {fn}")
    print("\n Classification Report:\n")
    print(report)

    return {
        'epochs': epochs,
        'batch_size': batch_size,
        'patience': early_stopping_patience,
        'accuracy': acc,
        'true_positives': tp,
        'true_negatives': tn,
        'false_positives': fp,
        'false_negatives': fn,
        'history': history,
        'report': report
    }


def plot_training_curves(history, title_suffix=""):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.title(f'Training vs Validation Accuracy {title_suffix}')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.title(f'Training vs Validation Loss {title_suffix}')
    plt.legend()

    plt.tight_layout()
    plt.show()


def compare_models():
    configs = [
        {'epochs': 50, 'batch_size': 16, 'early_stopping_patience': 10},
        {'epochs': 50, 'batch_size': 32, 'early_stopping_patience': 10},
        {'epochs': 50, 'batch_size': 16, 'early_stopping_patience': 3},
        {'epochs': 50, 'batch_size': 32, 'early_stopping_patience': 3},
        {'epochs': 20, 'batch_size': 32, 'early_stopping_patience': 3},
        {'epochs': 20, 'batch_size': 16, 'early_stopping_patience': 3},
        {'epochs': 20, 'batch_size': 64, 'early_stopping_patience': 3},
        {'epochs': 200, 'batch_size': 8,  'early_stopping_patience': 5},
        {'epochs': 200, 'batch_size': 16, 'early_stopping_patience': 5},
        {'epochs': 200, 'batch_size': 16, 'early_stopping_patience': 20}
    ]

    results = []
    for i, cfg in enumerate(configs, 1):
        print(f"\nTraining model {i}/{len(configs)} with config: {cfg}")
        result = train_bird_cnn(**cfg)
        results.append(result)
        plot_training_curves(result['history'], title_suffix=f"(epochs={cfg['epochs']}, batch_size={cfg['batch_size']})")

    df = pd.DataFrame(results)
    df = df[['epochs', 'batch_size', 'patience', 'accuracy',
             'true_positives', 'true_negatives',
             'false_positives', 'false_negatives', 'report', 'history']]

    print("\n Model Comparison Table:\n")
    print(df.drop(columns=['history', 'report']).to_string(index=False))

    df.drop(columns=['history', 'report']).to_csv("cnn_model_comparison.csv", index=False)
    print(" Saved model comparison table to cnn_model_comparison.csv")

    return df


## Script Execution 
if __name__ == "__main__":
    compare_models()
```
